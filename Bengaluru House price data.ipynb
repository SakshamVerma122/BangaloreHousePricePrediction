```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Bengaluru_House_Data.csv")
```

```python
data.sample(5)
```

```python
data.info()
```

```python
data.shape
```

```python
# value_counts
for i in data.columns:
    print(f"********************{i}**************************")
    print(data[i].value_counts())
    print(f">>>Number of unique elements: {len(data[i].unique())}<<<")
    print("***************************************************\n"*2)
```

```python
# isna on every column

numberCols = data.shape[0]
print("Percentages % of missing values")
for i in data.columns:
    print(f"{i}  {(data[i].isna().sum()/numberCols)*100}%")
```

From the above we can conclude that approximately 41% of the values are empty hence we must drop society column
